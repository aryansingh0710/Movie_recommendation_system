```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data= pd.read_csv('movies.csv')
movie_data
```

```python
#### selecting the relevent feature for recommondation  
selected_feature =['genres','keywords','tagline','cast','director']
print(selected_feature)
```

```python
for feature in selected_feature:
    movie_data[feature]= movie_data[feature].fillna('')
```

```python
##cobinig all the 5 features
combined_feature=movie_data['genres']+''+movie_data['cast']+''+movie_data['director']
```

```python
combined_feature
```

```python
#converting the text data into feature vectors 
vectorize = TfidfVectorizer()
```

```python
feature_vectore = vectorize.fit_transform(combined_feature)
print(feature_vectore)
```

```python
#getting the similrity score using cosine sisimilarity
similarity = cosine_similarity(feature_vectore)
print(similarity)
```

```python
similarity.shape
```

```python
movie_name= input('Enter your favourite movie name ')
```

```python
#creating a list with all the movie name given in dataset
list_of_all_title= movie_data['title'].tolist()
list_of_all_title
```

```python
# finding the close match for the movie name given by the user 
find_close_match = difflib.get_close_matches(movie_name,list_of_all_title)
find_close_match
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_match = movie_data[movie_data.title == close_match]['index'].values[0]
index_of_match
```

```python
similarity_score = list(enumerate(similarity[index_of_match]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score,key=lambda x:x[1],reverse=True)
sorted_similar_movies
```

```python
#print the name of similar movies based of the index 
print('movie suggested for you: /n')
i =1 
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index= movie_data[movie_data.index==index]['title'].values[0]
    if (i<30):
        print(i,'.',title_from_index)
        i+=1
```
